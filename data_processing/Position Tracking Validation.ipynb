# Data Visualization for Position Tracking Verification

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
input_data = np.loadtxt("D:\FYP_ML\Data\Verification_new_46.txt", delimiter = ',')
```

```python
timestamp = input_data[:,0]
initial_timestamp = timestamp[0]
timestamp = timestamp - initial_timestamp
timestamp
```

```python
# Data for plotting
plt.figure(figsize=(20, 10))

plt.plot(((timestamp*20*1e-8-0.2)*3), input_data[:,1]*0.005*10+5.5, c = 'r', label='Tracked Surface Variation',linewidth = 7)
plt.plot(((timestamp*20*1e-8-0.2)*3), input_data[:,2]*0.005+5.5, c = 'b', label='Reproduced Surface Variation',linewidth = 7)

plt.xlabel('Time (s)')
plt.ylabel('Position (mm)')
plt.title('Position vs time')

plt.xlim(0,10.5)
#plt.ylim(0,15)

plt.legend()
plt.grid()
plt.show()
```
